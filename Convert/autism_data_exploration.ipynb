```python
# Clean the 'age' column by removing invalid entries such as '?' and ages like '383'
# First, convert the column to numeric and handle errors by setting invalid parsing to NaN
df['age_clean'] = pd.to_numeric(df['age'], errors='coerce')

# Now that we have cleaned the age column, let's drop the NaN values and find the outliers
valid_age_data = df['age_clean'].dropna()

# Calculate the IQR for the age column
Q1_age = valid_age_data.quantile(0.25)
Q3_age = valid_age_data.quantile(0.75)
IQR_age = Q3_age - Q1_age

# Identify outliers in the age data
age_outliers = valid_age_data[(valid_age_data < (Q1_age - 1.5 * IQR_age)) | (valid_age_data > (Q3_age + 1.5 * IQR_age))]

age_outliers
```
